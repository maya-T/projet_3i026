```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

# Importation de votre librairie iads:
# La ligne suivante permet de préciser le chemin d'accès à la librairie iads
import sys
sys.path.append('../')   # iads doit être dans le répertoire frère du répertoire courant !

# Importation de la librairie iads
import iads as iads

# importation de Classifiers
from iads import Classifiers as cl

# importation de utils
from iads import utils as ut

import graphviz as gv
```

```python
import importlib
importlib.reload(ut)

# data set train

labels = pd.read_csv("Bases_IADS_multiclasses/names.csv", delimiter=";").columns
dataframe = pd.read_csv("Bases_IADS_multiclasses/train_multi.csv", delimiter=";", names = labels, index_col=0)


dataframe.head(5)

names = np.array(dataframe.columns)
names = np.delete(names, names.shape[0]-1)

train_data_label = np.array(dataframe['label'])
train_data_desc= np.array(dataframe[names])

print(train_data_desc.shape)
print(train_data_label.shape)

# creer le dictionnaire de labels ( classe i --> 1, le reste --> -1)
dico={}
for i in range(10):
    chaine="classe"+str(i)+"VSall_label"
    dico[chaine]=np.zeros(train_data_label.shape)
cpt=0
for i in dico:
    dico[i][train_data_label == cpt] = 1
    for j in range(10):
        if(j!=cpt):
            dico[i][train_data_label==j] = -1
    cpt+=1

# data set test

dataframe2 = pd.read_csv("Bases_IADS_multiclasses/test_multi.csv", delimiter=";", names = labels, index_col=0)

test_data_label = np.array(dataframe2['label'])
test_data_desc= np.array(dataframe2[names])
```

```python
# utiliser un perceptron avec un learning rate de 0,01

c_perceptron = cl.ClassifierPerceptron(train_data_desc.shape[1], 0.01)
results = {}
scores = {}
for k in dico:
    
    c_perceptron.train(train_data_desc, dico[k])
    for i in range(test_data_desc.shape[0]):
        if i not in results.keys() :
            results[i] = []
            scores[i] = []
        results[i].append(c_perceptron.predict(test_data_desc[i, :]))
        scores[i].append(c_perceptron.score(test_data_desc[i, :]))
    c_perceptron.reset()
```

```python
cpt = 0
for i in range(test_data_desc.shape[0]):
    
    print(i," : ",results[i])
    print(i," : ", scores[i])
    res = np.array(results[i])
    scr = np.array(scores[i])
    args = np.argwhere(res == 1)
    args = np.reshape(args, args.shape[0])
    print(args)
    if args.shape[0] == 0:
        rand = np.random.choice(np.arange(10), 1)
        prediction = rand[0] 
    else:    
#         prediction =np.random.choice(args, 1)
        predict_scores_absolute = np.absolute(scr[args])
        index = np.argmax(predict_scores_absolute)
        prediction = args[index]
    
    print("-->",prediction )
    print("-->",test_data_label[i])
    if(prediction == test_data_label[i]):
        cpt = cpt + 1 

accuracy = cpt/test_data_desc.shape[0]
print("accuracy : ", accuracy)

```
